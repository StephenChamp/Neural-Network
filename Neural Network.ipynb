# Neural Network 

### Libraries

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV, cross_val_score, learning_curve, train_test_split,KFold
import tensorflow as tf
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.callbacks import EarlyStopping
import numpy as np
from sklearn import metrics
from keras.utils import np_utils
from keras.wrappers.scikit_learn import KerasRegressor
import math
```

### Import Data

```python
data = pd.read_excel("HW3.xlsx")
data1 = pd.read_excel("HW3.xlsx")
```

### Remove extra variables

```python
data = data.drop('Purchase', 1)
data = data.drop('Web order', 1)
data = data.drop('last_update_days_ago', 1)
data = data.drop('1st_update_days_ago', 1)
```

### Create training and test set

```python
x = data.iloc[:, 1:18]
y = data.iloc[:,20]


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Normalize Frequency

```python
minmax = MinMaxScaler()
minmax.fit(x_train)
x_train = minmax.transform(x_train)

minmax = MinMaxScaler()
minmax.fit(x_test)
x_test = minmax.transform(x_test)
```

### Lets search for optimal parameters for our neural network, lets start with 2 layers

```python
def create_model(activation='relu', nb_hidden=10):
    model = Sequential()
    model.add(Dense(nb_hidden, input_dim=17, activation=activation))
    model.add(Dense(nb_hidden, input_dim=17, activation=activation))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=["accuracy"])
    return model
```

```python
nb_hiddens = np.array([10,20,40,60])
activations = ['relu']
a= ['tanh', 'sigmoid']
```

```python
param_grid = dict(activation=activations, nb_hidden=nb_hiddens)
model = KerasRegressor(build_fn=create_model, epochs=500, batch_size=256, verbose=2, validation_split=0.2)

clf = GridSearchCV(estimator=model, param_grid=param_grid, cv=4)
res = clf.fit(x_train, y_train)
```

```python
print(res.best_score_, res.best_params_)
for i in res.grid_scores_:
    print(i)
```

### Looks like 40 is the best lets use that!

### Build the Model

```python
model = Sequential()

## How many layers to the network
model.add(Dense(40, input_dim = x_train.shape[1], activation= 'relu'))
model.add(Dense(40, input_dim = x_train.shape[1], activation= 'relu'))
#model.add(Dense(20, activation= 'relu'))

#How will the network be evaluated
model.add(Dense(1))


## optimizer most complicated... adam is default go to
## loss generally measure MSE with regression
model.compile(optimizer = 'adam', loss = 'mean_squared_error')

```

### Train the model

```python


early_stopping_monitor = EarlyStopping(patience=10)

## different verbose if you want different output while it trains
## epochs is how many times to go through data
## validation split is splitting training data up to test the model on the current parameters
## call back will stop the model if it goes too long without improving

model.fit(x_train, y_train, epochs = 500, verbose = 2, 
          validation_split=0.2, batch_size = 256, callbacks=[early_stopping_monitor])



```

### Model Performance

```python
predictions = model.predict(x_test)
IN = []
for x in predictions:
    IN.append(x[0])

a = pd.Series(IN)


y_test
b =list(range(0,400))
y_test.index = b
actual_predicted = pd.concat((y_test, a), axis = 1)
actual_predicted.columns = ["Actual", "Predicted"]
actual_predicted['diff'] = (actual_predicted["Predicted"] - actual_predicted["Actual"]).round()
actual_predicted["% diff"] = actual_predicted["diff"] / actual_predicted["Actual"] 
actual_predicted["% diff"] = actual_predicted["% diff"].fillna(0)
avg_error = actual_predicted['diff'].mean()
absolute_error = actual_predicted['diff'].abs().mean()
squared = (actual_predicted['diff'] * actual_predicted['diff']).sum()
divided = squared/ actual_predicted["diff"].count()
RMSE = math.sqrt(divided)
percent_error = actual_predicted["% diff"].mean()
actual_predicted
```

```python
actual_predicted
absolute_error
RMSE
print("Neural Network Error Terms")
print("")
print("Avg Percent Error = %{0:.4f}".format(percent_error))
print("Avg Error = {0:.4f}".format(avg_error))
print("Avg Absolute Error = {0:.4f}".format(absolute_error))
print("RMSE = {0:.4f}".format(RMSE))
```
